```python
import numpy as np
```

```python
ratio = 1000
mlinestring = geom.MultiLineString(linestrings=[
    linestring.scale(freq=1/ratio, center=(0, 0)).rotate(angle, origin=linestring.end).scale(freq=ratio, center=(0, 0))
    for angle in np.linspace(0, 2 * np.pi, 60)
])
```

```python
_, ax = plt.subplots(figsize=(15, 5))
spec.plot(ax=ax)
mlinestring.plot(ax=ax, linewidth=2, color='red')
ax.set_ylim(0, spec._get_max());
```

```python
spec.to_mask(mlinestring).plot()
```

```python
spec.to_mask(mlinestring.buffer(time=0.05, freq=50)).plot(media=True, mask=False)
```
